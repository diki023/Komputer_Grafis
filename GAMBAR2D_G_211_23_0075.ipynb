```python
import numpy as np
import matplotlib.pyplot as plt

# Mengatur batas sumbu
plt.axis([-75, 75, 50, -50])

# Menyalakan sumbu
plt.axis('on')

# Menampilkan grid
plt.grid(True)

# Menggambar panah di sumbu x
plt.arrow(0, 0, 60, 0, head_length=4, head_width=3, color='k')

# Menggambar panah di sumbu y
plt.arrow(0, 0, 0, 45, head_length=4, head_width=3, color='k')

# Menambahkan label sumbu
plt.text(58, -3, 'x')
plt.text(-5, 44, 'y')

# Deklarasi variabel
a = 40
b = 20

# Inisialisasi variabel sudut
p1 = 0
p2 = 180 * np.pi / 180

# Diferensial sudut
dp = 0.2 * np.pi / 180

# Inisialisasi variabel untuk menyimpan koordinat
xplast = a
yplast = 0

# Perulangan untuk menghitung koordinat
for p in np.arange(p1, p2, dp):
    # Menghitung koordinat x dan y
    xp = np.abs(a * b * (b * b + a * a * (np.tan(p)) ** 2.) ** -.5)
    yp = np.abs(a * b * (a * a + b * b / (np.tan(p) ** 2.)) ** -.5)

    # Menggambar garis untuk sudut positif
    if p > np.pi / 2:
        xp = -xp

    plt.plot([xplast, xp], [yplast, yp], color='r')
    plt.plot([xplast, xp], [-yplast, -yp], color='r')

    # Memperbarui nilai variabel untuk iterasi selanjutnya
    xplast = xp
    yplast = yp

# Deklarasikan variabel
a = 20.
b = 40.

# Hitung koordinat x
xp1 = -a
xp2 = a

# Hitung selisih x
dx = .1

# Inisialisasi variabel untuk plot
xplast = -a
yplast = 0

# Hitung koordinat y dan plot elips
for xp in np.arange(xp1, xp2, dx):
    yp = b * (1 - xp**2. / a**2.)**.5
    plt.plot([xplast, xp], [yplast, yp], linewidth=1, color='g')
    plt.plot([xplast, xp], [-yplast, -yp], linewidth=1, color='g')

    # Perbarui variabel untuk plot berikutnya
    xplast = xp
    yplast = yp

# Menentukan nilai a dan b
a = 5.
b = 15.

# Menentukan rentang sudut
p1 = 0
p2 = 180 * np.pi / 180

# Menentukan langkah sudut
dp = .2 * np.pi / 180

# Menghitung koordinat x dan y untuk setiap sudut
for p in np.arange(p1, p2, dp):
    xp = np.abs(a * b * (b * b + a * a * (np.tan(p))**2.)**-.5)
    yp = np.abs(a * b * (a * a + b * b / (np.tan(p)**2.))**-.5)

    # Membalik tanda xp untuk sudut yang lebih besar dari 90 derajat
    if p > np.pi / 2:
        xp = -xp

    # Menggambar elips biru
    plt.plot([xp, xp], [yp, -yp], linewidth=1, color='b')

# Menampilkan plot
plt.show()
```

```python
"""
DROTRECTANGLE
"""

import matplotlib.pyplot as plt
import numpy as np

plt.axis([-10, 150, 100, -10])
plt.axis('on')
# Mengatur grid
plt.grid(True)

# Menggambar sumbu X dan Y
plt.arrow(0, 0, 40, 0, head_length=4, head_width=2, color='b')
plt.arrow(0, 0, 0, 40, head_length=4, head_width=2, color='b')

# Memberi label sumbu X dan Y
plt.text(30, -3, 'Xg', color='b')
plt.text(-8, 34, 'Yg', color='b')

# Menentukan pusat rotasi
xc = 75
yc = 50

# Menggambar garis X dan Y melalui pusat rotasi
plt.plot([xc-40, xc+60], [yc, yc], linewidth=1, color='grey')
plt.plot([xc, xc], [yc-40, yc+45], linewidth=1, color='grey')

# Memberi label titik X dan Y
plt.text(127, 48, 'X')
plt.text(70, 90, 'Y')

# Menandai pusat rotasi
plt.scatter(xc, yc, s=20, color='k')
plt.text(70, 49, 'c')
def rotz(xp, yp, rz):

    c11 = np.cos(rz)
    c12 = -np.sin(rz)
    c21 = np.sin(rz)
    c22 = np.cos(rz)

    xpp = xp * c11 + yp * c12  # relative to xc,yc
    ypp = xp * c21 + yp * c22

    xg = xc + xpp  # --relative to xg,yg
    yg = yc + ypp

    return [xg, yg]


# plot unrotated rectangle

xp1 = -20
xp2 = +20
xp3 = +20
xp4 = -20
# Koordinat titik sudut persegi panjang dalam sistem X, Y

yp1 = -5
yp2 = -5
yp3 = 5
yp4 = 5

plt.text(50,45,'1') # -- label
plt.text(97,45, '2')
plt.text(97,57,'3')
plt.text(50,57,'4')

# Plot titik sudut persegi panjang
plt.scatter(xp1 + xc, yp1 + yc, s=10, color='k')
plt.scatter(xp2 + xc, yp2 + yc, s=10, color='k')
plt.scatter(xp3 + xc, yp3 + yc, s=10, color='k')
plt.scatter(xp4 + xc, yp4 + yc, s=10, color='k')

# Hitung koordinat titik sudut persegi panjang yang diputar
xg1 = xc + xp1
yg1 = yc + yp1
xg2 = xc + xp2
yg2 = yc + yp2
xg3 = xc + xp3
yg3 = yc + yp3
xg4 = xc + xp4
yg4 = yc + yp4
# Membuat data
xg=[xg1,xg2,xg3,xg4,xg1]
yg=[yg1,yg2,yg3,yg4,yg1]
# Menampilkan plot
plt.plot((xg), (yg), color='k')

# Memutar koordinat sudut persegi panjang
rz = 45
rz = rz * np.pi / 180

# Menentukan koordinat sudut pertama
xp = xp1
yp = yp1
[xg,yg]=rotz(xp,yp,rz)
[xg1,yg1]=[xg,yg]
xp=xp2
yp=yp2
[xg,yg]=rotz(xp,yp,rz)
[xg2,yg2]=[xg,yg]
xp=xp2
yp=yp2
[xg,yg]=rotz(xp,yp,rz)
[xg3,yg3]=[xg,yg]
# Mendefinisikan variabel
xp =xp3
yp =yp3

# Menghitung rotasi
[xg, yg] = rotz(xp, yp, rz)

# Menyimpan nilai rotasi
[xg3, yg3] = [xg, yg]

xp=xp4
yp=yp4
[xg,yg]=rotz(xp,yp,rz)
[xg4,yg4]=[xg,yg]
xg=[xg1,xg2,xg3,xg4,xg1]
yg=[yg1,yg2,yg3,yg4,yg1]

plt.plot(xg,yg,color='r')

plt.show()
```
