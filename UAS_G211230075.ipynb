```python
import pandas as pd

# Load the data
file_path = 'stock.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python
# Clean the data
data['Purchase Price'] = data['Purchase Price'].str.replace('[\$,]', '', regex=True).astype(float)
data['Current Price'] = data['Current Price'].str.replace('[\$,]', '', regex=True).astype(float)
data['Cost Basis'] = data['Cost Basis'].str.replace('[\$,]', '', regex=True).astype(float)
data['Market Value'] = data['Market Value'].str.replace('[\$,]', '', regex=True).astype(float)
data['Gain/Loss'] = data['Gain/Loss'].str.replace('[\$,]', '', regex=True).astype(float)
data['Dividend/share'] = data['Dividend/share'].str.replace('[\$,]', '', regex=True).astype(float)
data['Annual Yield'] = data['Annual Yield'].str.replace('%', '', regex=True).astype(float)

# Display the cleaned data
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# a. Distribusi data pada setiap atribut Purchase Price, Current Price dan Gain/Loss dengan grafik histogram
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(data['Purchase Price'], kde=True, ax=axes[0])
axes[0].set_title('Distribusi Purchase Price')

sns.histplot(data['Current Price'], kde=True, ax=axes[1])
axes[1].set_title('Distribusi Current Price')

sns.histplot(data['Gain/Loss'], kde=True, ax=axes[2])
axes[2].set_title('Distribusi Gain/Loss')

plt.show()
```

```python
# b. Atribut Stock Name dan Symbol yang memiliki Dividend/share 5 tertinggi
top_dividends = data.nlargest(5, 'Dividend/share')
plt.figure(figsize=(10, 5))
sns.barplot(x='Dividend/share', y='Stock Name', data=top_dividends, hue='Symbol')
plt.title('Top 5 Dividend/share')
plt.show()
```

```python
# c. Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi
top_market_value = data.nlargest(5, 'Market Value')
plt.figure(figsize=(10, 5))
sns.barplot(x='Market Value', y='Stock Name', data=top_market_value, hue='Symbol')
plt.title('Top 5 Market Value')
plt.show()
```

```python
# d. Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi
top_annual_yield = data.nlargest(7, 'Annual Yield')
plt.figure(figsize=(10, 5))
sns.barplot(x='Annual Yield', y='Stock Name', data=top_annual_yield, hue='Symbol')
plt.title('Top 7 Annual Yield')
plt.show()
```
